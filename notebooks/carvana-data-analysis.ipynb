```python
def get_mask(img_path):
    return img_path.split('.')[0] + '_mask.gif'
```

```python
def plot_image_mask(img_path):
    img = Image.open(os.path.join('train', img_path))
    mask = Image.open(os.path.join('train_masks', get_mask(img_path)))
    
    plt.figure(figsize=(20, 10))
    plt.subplot(1, 3, 1)
    plt.imshow(img)
    
    plt.subplot(1, 3, 2)
    plt.imshow(mask)
    
    plt.subplot(1, 3, 3)
    plt.imshow(img)
    plt.imshow(mask, alpha=0.5)
    plt.show()
```

```python
img_path = random.choice(images)
plot_image_mask(img_path)
```
